```python
def compute_sorori_shinzaemon(n_days=100): #n_days is no of days / default = 0
    num = 1 # no. of 1st grain
    total = 1 # total no. of grains in 1st day
    list_n_grains = [num]# list of grains received each day
    list_total_grains = [total]
    for n in range(n_days - 1):
        num = num*2
        list_n_grains.append(num)#list of grains received each day
        total+=num
        list_total_grains.append(total)#list of total grains received by day
    print("Number of grains in day",n_days,':',num)
    print("Total number of grains by",n_days,"days:",total)
    
    import matplotlib.pyplot as plt #function to draw line graph
    %matplotlib inline
    fig, ax = plt.subplots()
    no_of_days=[x for x in range(1, n_days+1)]
    ax.plot(no_of_days, list_n_grains,label = 'No. of grains per day', color='red',linewidth = '2')
    ax.plot(no_of_days, list_total_grains,label = 'Total no. of grains received by day')
    plt.title("Total Rice received")
    plt.xlabel("Number of days")
    plt.ylabel("Total Number of grains")
    plt.legend()
    plt.show()


compute_sorori_shinzaemon(30)# compute for 30 days
    
        
        
```

# Question 3:
One gram of rice has around 48 grains. So a cup of 180 grams has 8640 grains of rice.
The best way to maintain your weight in a healthy range while eating rice is to exercise portion control. It is advisable to limit your consumption of rice to 1 cup of cooked rice per day.
Based on this we take number of grains per person per day as 8640 grains. assuming that our class has 100 students.

```python
def days_to_live_with_rice(n_of_grains, n_of_people):
    no_grains_person = 8640
    no_of_days = n_of_grains // (n_of_people * no_grains_person)
    return no_of_days
print(days_to_live_with_rice(1329227995784915872903807060280344575, 100))

    
```
